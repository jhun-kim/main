```python
corona_df_2 = corona_df.sort_values('등록일시') #acending 쓸수 있음
corona_df_2
```

```python
corona_df_2["일일 확진자"] = (corona_df_2["확진자"] - 
                            corona_df_2["확진자"].shift()).fillna(0)
                            # 쉬프트는 행을 하나 줘서, 칸을 만든다. 
corona_df_2
```

```python
corona_df_2["일일사망자"] = (corona_df_2["사망자"]-
                            corona_df_2["사망자"].shift()).fillna(0)
                            #fillna는 결측값을 ()으로 만드는 함수
corona_df_2
```

```python
#결측값 = 값이 없는 값
```

```python
corona_df_2["일일 사망자"] = corona_df_2["사망자"].diff().fillna(0)
corona_df_2
```

```python
corona_df_2["일일 사망자"].value_counts(ascending=True)
# 카운트의 갯수로 true-오름차순 된다.)
#밸류값들의 갯수
```
